```python
import pandas as pd

# === Load dataset ===
df = pd.read_csv("default+of+credit+card+clients_tiwan/default of credit card clients.csv")

# === Remove metadata row if present ===
# If row 0 contains column descriptions (like 'LIMIT_BAL', 'SEX', etc.), drop it
if df.iloc[0].isnull().sum() < len(df.columns) // 2:  # crude check for metadata row
    df = df.drop(index=0).reset_index(drop=True)

# === Rename columns with descriptive names ===
column_mapping = {
    'Unnamed: 0': 'ID',
    'X1': 'Amount_of_Credit',
    'X2': 'Gender',
    'X3': 'Education_Level',
    'X4': 'Marital_Status',
    'X5': 'Age',
    'X6': 'Repayment_Status_Sep_2005',
    'X7': 'Repayment_Status_Aug_2005',
    'X8': 'Repayment_Status_Jul_2005',
    'X9': 'Repayment_Status_Jun_2005',
    'X10': 'Repayment_Status_May_2005',
    'X11': 'Repayment_Status_Apr_2005',
    'X12': 'Bill_Statement_Sep_2005',
    'X13': 'Bill_Statement_Aug_2005',
    'X14': 'Bill_Statement_Jul_2005',
    'X15': 'Bill_Statement_Jun_2005',
    'X16': 'Bill_Statement_May_2005',
    'X17': 'Bill_Statement_Apr_2005',
    'X18': 'Previous_Payment_Sep_2005',
    'X19': 'Previous_Payment_Aug_2005',
    'X20': 'Previous_Payment_Jul_2005',
    'X21': 'Previous_Payment_Jun_2005',
    'X22': 'Previous_Payment_May_2005',
    'X23': 'Previous_Payment_Apr_2005',
    'Y': 'Default_Payment_Next_Month'
}
df.rename(columns=column_mapping, inplace=True)

# === Ensure key columns are numeric ===
cols_to_convert = ['Default_Payment_Next_Month', 'Gender', 'Education_Level', 'Marital_Status'] + \
    [f'Repayment_Status_{month}_2005' for month in ['Sep', 'Aug', 'Jul', 'Jun', 'May', 'Apr']]
df[cols_to_convert] = df[cols_to_convert].apply(pd.to_numeric, errors='coerce')

# === Define value mappings ===
default_map = {1: 'Yes', 0: 'No'}
gender_map = {1: 'Male', 2: 'Female'}
education_map = {1: 'Graduate School', 2: 'University', 3: 'High School', 4: 'Others'}
marital_map = {1: 'Married', 2: 'Single', 3: 'Others'}
repayment_map = {
    -1: 'Paid Duly',
     0: 'No Consumption',
     1: 'Delay 1 Month',
     2: 'Delay 2 Months',
     3: 'Delay 3 Months',
     4: 'Delay 4 Months',
     5: 'Delay 5 Months',
     6: 'Delay 6 Months',
     7: 'Delay 7 Months',
     8: 'Delay 8 Months',
     9: 'Delay 9+ Months'
}

# === Apply value mappings directly to columns ===
df['Default_Payment_Next_Month'] = df['Default_Payment_Next_Month'].map(default_map)
df['Gender'] = df['Gender'].map(gender_map)
df['Education_Level'] = df['Education_Level'].map(education_map)
df['Marital_Status'] = df['Marital_Status'].map(marital_map)

for col in [f'Repayment_Status_{month}_2005' for month in ['Sep', 'Aug', 'Jul', 'Jun', 'May', 'Apr']]:
    df[col] = df[col].map(repayment_map)

# === Optional: Fill unmapped values with 'Unmapped' ===
#df = df.fillna('Unmapped')
```

```python
df.drop(df.columns[0], axis=1, inplace=True)
df
```

```python
df.to_csv('default+of+credit+card+clients_tiwan/default of credit card clients_with_information.csv', index=False)
```
